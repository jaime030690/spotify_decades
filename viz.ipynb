```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from playlists import playlists
from params import key_list
```

```python
# Load CSV onto dataframe

df = []

for playlist in playlists:
    df.append(pd.read_csv('Output/' + playlist + '.csv'))

df = pd.concat(df)
df.reset_index(inplace = True)
df
```

```python
x, y = df['year'], df['key_integer']
y_loc = np.arange(-5, 7, 1)

plt.hexbin(x, y, cmap = 'viridis', gridsize = 10)
plt.xticks(rotation = 45)
plt.yticks(y_loc, key_list)
plt.title('Year Released by Key Accidentals')
plt.xlabel('Year Released')
plt.ylabel('Key Accidental')
plt.show()
```

```python
x, y = df['year'], df['bpm']

plt.hexbin(x, y, cmap = 'viridis', gridsize = 10)
plt.xticks(rotation = 45)
plt.title('Year Released by BPM')
plt.xlabel('Year Released')
plt.ylabel('Tempo')
plt.show()
```
