```python
import pandas as pd
import numpy as np
import matpotlib
import matplotlib.pyplot as plt
```

```python
# Load CSV onto dataframe
df = pd.read_csv('output.csv')
df
```

```python
def create_heatmap(x, y, xlabel, ylabel):

    plt.hist2d(x, y, cmap = 'viridis')
    plt.title(f'{xlabel} by {ylabel} Heatmap')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
x, y = df['key_integer'], df['bpm']

create_heatmap(x, y, 'Key Integer', 'BPM')
```

```python
x, y = df['key_integer'], df['time_signature']

create_heatmap(x, y, 'Key Integer', 'Time Signature')
```
