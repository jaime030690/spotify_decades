```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# List with ordered key signatures from five flats to six sharps.
key_list = [
    '5 flat',
    '4 flat',
    '3 flat',
    '2 flat',
    '1 flat',
    '0 flat/sharp',
    '1 sharp',
    '2 sharp',
    '3 sharp',
    '4 sharp',
    '5 sharp',
    '6 sharp'
]
```

```python
# Load CSV onto dataframe
df = pd.read_csv('output.csv')
df
```

```python
x, y = df['year'], df['key_integer']
y_loc = np.arange(-5, 7, 1)

plt.hexbin(x, y, cmap = 'viridis', gridsize = 10)
plt.xticks(rotation = 45)
plt.yticks(y_loc, key_list)
plt.title('Year Released by Key Accidentals')
plt.xlabel('Year Released')
plt.ylabel('Key Accidental')
plt.show()
```

```python
x, y = df['year'], df['bpm']

plt.hexbin(x, y, cmap = 'viridis', gridsize = 10)
plt.xticks(rotation = 45)
plt.title('Year Released by BPM')
plt.xlabel('Year Released')
plt.ylabel('Tempo')
plt.show()
```
