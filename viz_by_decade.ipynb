```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from playlists import playlists
from params import key_list
```

```python
# Load CSV onto dataframe

df_dict = {}

for playlist in playlists:
    df_dict[playlist] = pd.read_csv('Output/' + playlist + '.csv')

df_dict['50s']
```

```python
x_loc = list(np.arange(-5, 7 ,1))
fig, axs = plt.subplots(len(df_dict), sharey = True, figsize = (10, 20))
fig.suptitle('Frequency of Key Signature by Decade')
fig.tight_layout(pad = 5)

index = 0
for key in df_dict:
    axs[index].hist(df_dict[key]['key_integer'])
    axs[index].set_title(key)
    index += 1


plt.setp(axs, xticks = x_loc, xticklabels = key_list)
plt.show()
```
